 * **Columns:**

# Converting datatype of release date column obejct to datetime

```python
movie['release_date'] = pd.to_datetime(movie['release date'], format = "%d-%m-%Y")
movie.drop('release date', axis=1,inplace=True)
```

# Number of movies released after 1990

```python
movie['Year'] = movie['release_date'].dt.year
movie.head()
```

```python
years = movie['Year'].value_counts()
years
```

```python
movie[movie['Year']>1990]['Year'].value_counts()
```

## Insights
- We can observe that the highest number of movies with 355 in total have been released in 1996. Also, we can see that the count of the number of movies released in each year is increasing except for 1997 and 1998. So it can be said that, overall there is an upward trend in the number of movie released between the year 1991 and 1998.
- The top five 'years' with the highest number of movies released are '1996', '1997', '1995', '1994', and '1993'.

# Most liked genres by users

```python
df_merge = movie.merge(rating, on = 'movie id', how = 'inner')
df_merge.head()
```

```python
mean_rating = pd.DataFrame()

# Iterating through the columns of df_merge to extract and store the ratings for each genre
for i in genres:
  mean_rating[i] = np.where(df_merge[i]==1,df_merge['rating'],np.nan)

# printing the newly created dataframe
mean_rating.head()
```

```python
mean_rating.mean(axis=0).sort_values(ascending = False)
```

## Insights
- 'Film-Noir' genre is having the highest average ratings with an average rating of 3.92 followed by 'War' genre with rating of 3.81. So, the majority of the population tends to like the 'Film-Noir' and 'War' movies compared to any other movies genres.
- Although, the total number of movies in 'Film-Noir' and 'Fantasy' are almost equal, 'Fantasy' genre is having the lowest average ratings with an average rating of 3.21.
- Among 18 genres, only 5 genres have received a rating below 3.5 which means around 72% of genres are rated above average

Till now we have done analyses on the genres and movies and generated some insights.

# Most preferred movies by users

Let's start by checking the top rated movies by the users. As we already know that `movie` and `ratings` datasets has the information related to the ratings of the movies.
Therefore, we can merge both the datasets on the basis of common variable that is `movie_id` and get the average ratings for all the movies. Since we have already merged these datasets before, we can call it again to get the values

```python
avg_rating = df_merge.groupby(['movie title']).mean()[['rating']].reset_index() 

avg_rating.head()
```

```python
avg_rating[avg_rating['rating']==1]
```

```python
avg_rating[avg_rating['rating']==5]
```

## Insights
- We can observe movies like Great Day in Harlem, A, Prefontaine etc are top rated movies
- Movies like Shadow of Angels (Schatten der Engel), Power 98, etc are least rated
- There are 10 movies which have been rated as 5.0

# Relationship between demographic details of users and ratings of the movies

```python
df_merge_all = df_merge.merge(user, on = 'user id', how = 'inner')
```

```python
user['gender'].value_counts()
```

```python
df_merge_all.groupby('gender').mean()['rating']
```

```python
df_merge_all.groupby('occupation').mean()['rating'].sort_values(ascending = False)
```

```python
df_merge_all.groupby(['occupation','gender']).mean()['rating'].sort_values(ascending = False)
```

## Insights
- Around 71% of users are Males while only 29% users are females
- The average rating given by Males and Females is almost equal with a value of around 3.53 and it can be said that the ratings given by Males and Females does not differ significantly.
- Users who are not working have given highest ratings as compared to users who are working professionals.
- Healthcare workers have given lowest ratings on an average
- The low ratings from the healthcare sector is majorly driven by female workers
